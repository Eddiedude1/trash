```python
import pandas as pd
import numpy as np
import json
import pprint
```

```python
# read in the 10 percente set
df10 = pd.read_csv('./data/Iowa_Liquor_sales_sample_10pct.csv')
```

```python
# try to read in the 100% set
df = pd.read_csv('./data/iowa_liquor_sales_proj_2.csv')
```

## See if there is a difference between 100% and 10% datasets

```python
### Find out what columns exist in the 100% dataset 
### that don't in the 10% dataset
# make colulmns into dictionary
df10_dict = { 'cols': [] }
for c in df10.columns.tolist():
    df10_dict['cols'].append(c)

df_dict = { 'cols': [] }
for c in df.columns.tolist():
    df_dict['cols'].append(c)
    
# make dict into dfs
df10_cols = pd.DataFrame(df10_dict)
df_cols = pd.DataFrame(df_dict)

# create a new column to hold the cols data, 
# so I can use set_index and retain a value col
df_cols['df_100_cols'] = df_cols['cols']
df10_cols['df_10_cols'] = df10_cols['cols']

# joining the 100% to the 10%
# df10_cols.join(df_cols.set_index('cols'), on='cols', how='outer', lsuffix='1', rsuffix='2')

# joining the 10% to the 100%
missing = df_cols.join(df10_cols.set_index('cols'), on='cols', how='outer', lsuffix='1', rsuffix='2')
```

#### Looks like we have 5 columns that are in the 100% dataset that aren't in the 10% dataset

```python
missing[ missing['df_100_cols'].isnull() | missing['df_10_cols'].isnull() ].iloc[:,1:]['df_100_cols'].tolist()
```
